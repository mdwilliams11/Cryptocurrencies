```python
import pandas as pd
```

```python
# read in the csv
file_path = "Resources/iris.csv"
iris_df = pd.read_csv(file_path)
iris_df.head()
```

```python
# drop the class field
new_iris_df = iris_df.drop(["class"], axis=1)
new_iris_df.head()
```

```python
# reorder the columns
column_names = ["sepal_length", "petal_length", "sepal_width", "petal_width"]
new_iris_df = new_iris_df.reindex(columns=column_names)
new_iris_df.head()
```

```python
# export cleaned csv
output_file_path = "Resources/new_iris_data.csv"
new_iris_df.to_csv(output_file_path, index=False)
```

```python
# 18.2.3
# read in the csv
file_path = "Resources/shopping_data.csv"
df_shopping = pd.read_csv(file_path, encoding="ISO-8859-1")
df_shopping.head()
```

```python
# Columns
df_shopping.columns
```

```python
# Columns
df_shopping.dtypes
```

```python
# find if there's null values
for column in df_shopping.columns:
    print(f"Column {column} has {df_shopping[column].isnull().sum()} null values")
```

```python
# Drop null rows
df_shopping = df_shopping.dropna()
df_shopping.head()
```

```python
# find duplicate entries
print(f"Duplicate entries: {df_shopping.duplicated().sum()}")
```

```python
df_shopping.head()
```

```python
df_shopping.drop(columns=["CustomerID"], inplace=True)
df_shopping.head()
```

```python
# transform string column
def change_string(member):
    if member == "Yes":
        return 1
    else:
        return 0

df_shopping["Card Member"] = df_shopping["Card Member"].apply(change_string)
df_shopping.head()
```

```python
# transform the annual income to rescale it
df_shopping["Annual Income"] = df_shopping["Annual Income"] / 1000
df_shopping.head()
```

```python
df_shopping = df_shopping.rename({"Card Member": "card_member", "Age": "age","Annual Income": "annual_income", "Spending Score (1-100)": "spending_score"}, axis=1)
df_shopping.head()
```
